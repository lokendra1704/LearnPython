```python
import pickle
import os

class student:
    def __init__(self,i):
        self.id = int(input(f'Enter roll number of {i}th Student'))
        self.name = input(f'Enter name of {i}th Student')
    def Display(self):
        print('-------------------------------',f'Student ID: {self.id}\nStudent Name: {self.name}\n-------------------------------',sep='\n')

n = int(input("Enter no of Student"))
with open('lalit.bin','wb') as file:
    for i in range(n):
        s = student(i+1)
        pickle.dump(s,file)

with open('lalit.bin','rb') as file:
    while True:
        try:
            s = pickle.load(file)
            s.Display()
        except Exception as e:
            break
            
with open('lalit.bin','rb') as file:
    searchid = int(input("Searching: Enter ID of Student"))
    while True:
        try:
            print(file.tell())
            s = pickle.load(file)
            if s.id == searchid:
                print("Student Found")
                break
        except Exception as e:
            print('Not found')
            break

file = open('lalit.bin','rb+')
file1 = open('temp.bin','wb')
changeid = int(input("Change Data: Enter ID of Student"))
while True:
    try:
        t = pickle.load(file)
        if t.id == changeid:
            t.id = int(input('Enter new roll'))
            t.name = input('Enter new name')
        pickle.dump(t,file1)
    except:
        print('Roll Number not found')
        break
file.close()
file1.close()
os.remove('lalit.bin')
os.rename('temp.bin','lalit.bin')

with open('lalit.bin','rb') as file:
    while True:
        try:
            s = pickle.load(file)
            s.Display()
        except Exception as e:
            break
```
